```python
attributes_poly = PolynomialFeatures().fit_transform(attributes)
```

```python
model_poly = LinearRegression()
```

```python
# The model needs more time so this means the model makes more calculations
# which is not always good
model_poly.fit(attributes_poly, targets)
```

```python
model_poly.score(attributes_poly, targets)
```

High bias --> 1) Less data
              2) More features = more information

Over-fitting
Very big or little weight coeficients mean that with a very small difference in the input we can create different output. This is a model which is chaotic and it overfits.
Very big coeficients -> something is not right

```python
np.linalg.norm(model.coef_, ord = 2) # square error X^2
```

```python
np.linalg.norm(model.coef_, ord = 1) # abs error |X|
```

```python
np.linalg.norm(model.coef_, ord = 0) # len(x); the length of 0 elements
```

```python
np.linalg.norm(model.coef_, ord = np.inf) # abs(max(x))
```

```python
# Ridge does L2, Lasso does L1 and Elastic_Net does both

ridge_quad = Ridge()
```

```python
ridge_quad.fit(attributes_poly, targets)
```

```python
ridge_quad.score(attributes_poly, targets)
```

Let's say the previous model has overfitted and plays with the data too much, then the ridge score should be lower.

The ridge model is better because the previous one does more than I want.
The previous model does not generalize well and is too good so that I do a regularisation.

```python
np.linalg.norm(model_poly.coef_)
```

```python
 # the sum of the squares of the coeficients is less now
np.linalg.norm(ridge_quad.coef_)
```

With this ridge model I have told the model not to do the coeficients big but I have received a function which is more predictable. The perfect model does not apply the metrics I pass in the begininng.

```python
lasso_quad = Lasso()
```

```python
lasso_quad.fit(attributes_poly, targets)
```

```python
np.linalg.norm(lasso_quad.coef_)
```

Lasso aligns some coeficients to 0 while Ridge reduce the coeficients proportionally. Some coeficients become 0 too if there are linearly dependent features.

```python
lasso_quad.score(attributes_poly, targets)
```
