### Machine Learning Part

```python
# Remove the column that will be predicted
# Attributes are the features that the model will use
attributes = stockx_data.drop(columns = "sale_price")
```

```python
# In classification targets are called labels
# Sale prices will be threated as target
# The name should not be matched with the meaning of the feature (for example targer)
 # but with what is appropriate according its application 
    # In this case, the application is not target but sale price so "sale_price" is more descriptive
targets = stockx_data.sale_price
```

```python
model = LinearRegression()
```

```python
model.fit(attributes, targets)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# estimator is hyperparameter and by default is a Linear Regression but can be some other regressor
ransac = RANSACRegressor(estimator = LinearRegression()) # meta-learning
```

```python
ransac.fit(attributes, targets)
```

```python
ransac.inlier_mask_.sum() / len(ransac.inlier_mask_)
```

```python
# estimator is hyperparameter and by default is a Linear Regression but can be some other regressor
ransac = RANSACRegressor(estimator = LinearRegression(), min_samples = 0.8) # meta-learning
# uses at least 80% of the data for one algorithm
```

```python
ransac.fit(attributes, targets)
```

```python
ransac.inlier_mask_.sum() / len(ransac.inlier_mask_)
```

```python
ransac.score(attributes, targets)
```

Algorithm got worse. The alogithm above has given 63% of the data.

```python
inliers = attributes.loc[ransac.inlier_mask_]
inlier_targets = targets[ransac.inlier_mask_]
```

```python
outliers = attributes.loc[~ransac.inlier_mask_]
outlier_targets = targets[~ransac.inlier_mask_]
```

```python
len(inlier_targets)
```

```python
# Algorithm knows the data - core inliers; many models have worked over them
# Works well on data it knows but it is a bias
ransac.score(inliers, inlier_targets)
```

```python
# The algorithm does not know them
ransac.score(outliers, outlier_targets)
```

```python
# The regression I have is weak. How to fix it?
# Give less data and more features so that the model can focus on them.
# More features = more flexibility

# If the model is good, then it can be given more data and less features
```

```python
outliers.shoe_size.mean()
```

```python
inliers.shoe_size.mean()
```

```python
poly_transformer = PolynomialFeatures()
```

```python
poly_transformer.fit(attributes)
```

```python
attributes.shape
```

```python
poly_transformer.n_output_features_
```

```python
poly_transformer.n_features_in_
```

```python
attributes_squared = poly_transformer.transform(attributes)
```

```python
quadratic_model = LinearRegression()
```

```python
quadratic_model.fit(attributes_squared, targets)
```

```python
# This model is better because it has more features
quadratic_model.score(attributes_squared, targets)
```

WARNING: The models here are being assessed only on the data which the model has used for training so that is not the perfectly correct way of estimation.

```python
# Polynomial features
list(zip(poly_transformer.get_feature_names_out(), quadratic_model.coef_))
```
